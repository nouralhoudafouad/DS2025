```python
# -*- coding: utf-8 -*-
# Notebook : Analyse du fichier ADANIPORTS (2).csv
# Collez ce code dans Google Colab. Si besoin, uploadez le fichier via Files -> Upload.

# -----------------------
# 0) Upload si besoin (décommentez si nécessaire)
# -----------------------
from google.colab import files
uploaded = files.upload()

# -----------------------
# 1) Imports
# -----------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import DateFormatter
%matplotlib inline

plt.rcParams['figure.figsize'] = (12,5)
pd.set_option('display.max_columns', 200)

# -----------------------
# 2) Lecture du fichier
# -----------------------
csv_path = "ADANIPORTS (2).csv"   # <- changez si vous avez un autre chemin
df = pd.read_csv(csv_path)

print("Shape:", df.shape)
display(df.head(6))

# -----------------------
# 3) Nettoyage & parsing
# -----------------------
df.columns = [c.strip() for c in df.columns]

# Convertir Date en datetime
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Nettoyer colonnes numériques (retirer % et séparateurs de milliers)
for c in df.columns:
    if df[c].dtype == object and c not in ['Date','Symbol','Series']:
        df[c] = df[c].astype(str).str.replace('%','').str.replace(',','').str.strip()
    # tenter conversion
    if c not in ['Date','Symbol','Series']:
        df[c] = pd.to_numeric(df[c], errors='coerce')

print("\nTypes et valeurs manquantes par colonne:")
display(df.dtypes)
display(df.isna().sum())

# -----------------------
# 4) Résumé général
# -----------------------
date_min = df['Date'].min() if 'Date' in df.columns else None
date_max = df['Date'].max() if 'Date' in df.columns else None
n_rows, n_cols = df.shape
n_symbols = df['Symbol'].nunique() if 'Symbol' in df.columns else 0

print(f"Enregistrements: {n_rows}, Colonnes: {n_cols}, Symbols uniques: {n_symbols}")
if date_min is not None and date_max is not None:
    print(f"Période: {date_min.date()} → {date_max.date()}")

# Top symboles par volume et par turnover
if 'Volume' in df.columns:
    top_vol = df.groupby('Symbol')['Volume'].sum().sort_values(ascending=False).head(10)
    print("\nTop symboles par Volume (total):")
    display(top_vol)
if 'Turnover' in df.columns:
    top_turn = df.groupby('Symbol')['Turnover'].sum().sort_values(ascending=False).head(10)
    print("\nTop symboles par Turnover (total):")
    display(top_turn)

# -----------------------
# 5) Visualisations principales
# -----------------------

def plot_time_series(sym, col='Close', window_ma=30):
    d = df[df['Symbol']==sym].sort_values('Date')
    if d.empty:
        print("Aucun enregistrement pour", sym)
        return
    d = d.set_index('Date').asfreq('D')
    fig, ax = plt.subplots()
    ax.plot(d.index, d[col], linewidth=1, label=f'{sym} {col}')
    if col in d.columns:
        ax.plot(d.index, d[col].rolling(window=window_ma, min_periods=1).mean(),
                linewidth=1, linestyle='--', label=f'MA{window_ma}')
    ax.set_title(f'{sym} — {col} (période: {date_min.date()}→{date_max.date()})')
    ax.set_xlabel('Date')
    ax.set_ylabel(col)
    ax.legend()
    ax.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

symbols = df['Symbol'].unique().tolist() if 'Symbol' in df.columns else []
print("Symbols trouvés:", symbols)
for s in symbols:
    if 'Close' in df.columns:
        plot_time_series(s, col='Close', window_ma=30)
    if 'Volume' in df.columns:
        plot_time_series(s, col='Volume', window_ma=30)

# Histogramme %Deliverble si présent
if '%Deliverble' in df.columns:
    plt.figure()
    sns.histplot(df['%Deliverble'].dropna(), bins=40, kde=True)
    plt.title("Distribution de % Deliverable")
    plt.xlabel("% Deliverable")
    plt.tight_layout()
    plt.show()

# Corrélation entre colonnes numériques
num_df = df.select_dtypes(include=[np.number]).copy()
if num_df.shape[1] > 1:
    corr = num_df.corr()
    plt.figure(figsize=(10,8))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', center=0)
    plt.title("Matrice de corrélation (colonnes numériques)")
    plt.tight_layout()
    plt.show()

# Rendements journaliers (Close) par symbole
if 'Close' in df.columns:
    for s in symbols:
        d = df[df['Symbol']==s].sort_values('Date').set_index('Date').resample('D').last()
        if 'Close' in d.columns:
            d['ret'] = d['Close'].pct_change()
            plt.figure()
            sns.histplot(d['ret'].dropna(), bins=60, kde=True)
            plt.title(f'Distribution des rendements journaliers — {s}')
            plt.xlabel('Rendement journalier')
            plt.tight_layout()
            plt.show()

# Top journées par Volume / Turnover
if 'Volume' in df.columns:
    top_days_vol = df.sort_values('Volume', ascending=False).head(10)[['Date','Symbol','Volume','Close']]
    print("\nTop 10 journées par Volume:")
    display(top_days_vol)
if 'Turnover' in df.columns:
    top_days_turn = df.sort_values('Turnover', ascending=False).head(10)[['Date','Symbol','Turnover','Close']]
    print("\nTop 10 journées par Turnover:")
    display(top_days_turn)

# -----------------------
# 6) Analyses additionnelles & export
# -----------------------
summary_by_symbol = df.groupby('Symbol').agg(
    first_date = ('Date','min'),
    last_date  = ('Date','max'),
    days        = ('Date', lambda x: x.nunique()),
    total_volume = ('Volume','sum'),
    mean_close   = ('Close','mean'),
    median_close = ('Close','median'),
    avg_deliverable_pct = ('%Deliverble','mean')
).reset_index()
print("Résumé par symbole:")
display(summary_by_symbol)

summary_by_symbol.to_csv("resume_par_symbol_ADANIPORTS_2.csv", index=False)
print("Fichier 'resume_par_symbol_ADANIPORTS_2.csv' créé dans l'environnement.")
```
